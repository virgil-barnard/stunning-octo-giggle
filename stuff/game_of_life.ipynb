```python
# matplotlib qt
import numpy as np
import time
import matplotlib.pyplot as plt
from IPython.display import clear_output

class Game:
    def __init__(self, width, height, sleep_timer):
        self.width = width
        self.height = height
        self.sleep_timer = sleep_timer
        self.grid = np.array([[Cell(x, y) for y in range(height)] for x in range(width)])

    def randomize(self):
        for row in self.grid:
            for cell in row:
                cell.alive = np.random.choice([False, True])

    def plot_grid(self):
        alive_x = [cell.x for row in self.grid for cell in row if cell.alive]
        alive_y = [cell.y for row in self.grid for cell in row if cell.alive]
        plt.figure(figsize=(20, 20))  # Adjust figure size here
        plt.scatter(alive_x, alive_y, color='k')
        plt.xlim(0, self.width)
        plt.ylim(0, self.height)
        plt.show()

    def count_neighbours(self, x, y):
        count = 0
        for i in range(-1, 2):
            for j in range(-1, 2):
                if i == 0 and j == 0:
                    continue
                elif (x + i < 0 or y + j < 0 or x + i >= self.width or y + j >= self.height):
                    continue
                elif self.grid[x + i][y + j].alive:
                    count += 1
        return count

    def update(self):
        for row in self.grid:
            for cell in row:
                neighbours = self.count_neighbours(cell.x, cell.y)
                if cell.alive:
                    if neighbours < 2 or neighbours > 3:
                        cell.next_state = False
                else:
                    if neighbours == 3:
                        cell.next_state = True

        for row in self.grid:
            for cell in row:
                cell.alive = cell.next_state

    def run(self, max_steps):
        self.randomize()
        for _ in range(max_steps):
            clear_output(wait=True) 
            self.plot_grid()
            self.update()
            time.sleep(self.sleep_timer)
```

```python
import numpy as np
import random
from matplotlib.colors import ListedColormap
from IPython.display import clear_output
import time
import matplotlib.pyplot as plt


class NeuralNetwork:
    def __init__(self, genome):
        self.connections = {}
        self.weights = {}
        self.sensory_nodes = ['up', 'down', 'left', 'right']

        self.build_network(genome)

    def build_network(self, genome):
        for gene in genome:
            source, sink, weight = gene
            self.connections[sink] = source
            self.weights[sink] = weight

    def get_input(self, creature, direction):
        x, y = creature.x, creature.y
        if direction == 'up':
            return y
        elif direction == 'down':
            return creature.world.height - 1 - y
        elif direction == 'left':
            return x
        elif direction == 'right':
            return creature.world.width - 1 - x

    def activate(self, creature):
        inputs = [self.get_input(creature, direction) for direction in self.sensory_nodes]

        outputs = {}
        for sink, source in self.connections.items():
            if source in inputs:
                index = inputs.index(source)
                weight = self.weights[sink]
                output = outputs.get(sink, 0)
                outputs[sink] = output + weight * inputs[index]

        return outputs


class Creature:
    def __init__(self, x, y, genome, world):
        self.x = x
        self.y = y
        self.genome = genome
        self.world = world
        self.brain = NeuralNetwork(genome)

    def mutate(self, mutation_rate):
        new_genome = []
        for gene in self.genome:
            if random.random() < mutation_rate:
                new_weight = random.uniform(-1, 1)
                new_gene = (gene[0], gene[1], new_weight)
            else:
                new_gene = gene
            new_genome.append(new_gene)
        return new_genome


class World:
    def __init__(self, width, height, sleep_timer=0.1, mutation_rate=0.5, harvesting_period=10, max_population=500):
        self.width = width
        self.height = height
        self.grid = np.empty((width, height), dtype=object)
        self.sleep_timer = sleep_timer
        self.mutation_rate = mutation_rate
        self.harvesting_period = harvesting_period
        self.max_population = max_population

    def populate(self, n_creatures):
        for _ in range(n_creatures):
            x = random.randint(0, self.width - 1)
            y = random.randint(0, self.height - 1)
            genome = self.generate_genome()
            self.grid[x, y] = Creature(x, y, genome, self)

    def generate_genome(self):
        genome_length = random.randint(5, 10)  # Random genome length between 5 and 10
        genome = []
        for _ in range(genome_length):
            source = random.choice(range(4))  # Randomly select a sensory node as the source
            sink = random.randint(4, 7)  # Randomly select an internal node as the sink
            weight = random.uniform(-1, 1)  # Random weight between -1 and 1
            gene = (source, sink, weight)
            genome.append(gene)
        return genome

    def update(self, iteration):
        # Movement phase
        for x in range(self.width):
            for y in range(self.height):
                creature = self.grid[x, y]
                if creature is not None:
                    outputs = creature.brain.activate(creature)
#                     print(outputs)
                    if outputs:
                        action = max(outputs.items(), key=lambda x: x[1])[0]
                    else:
                        action = "None"
#                         action = random.choice(['up', 'down', 'left', 'right'])

                    new_x = x
                    new_y = y
                    if action == 'up' and y < self.height - 1:
                        new_y += 1
                    elif action == 'down' and y > 0:
                        new_y -= 1
                    elif action == 'left' and x > 0:
                        new_x -= 1
                    elif action == 'right' and x < self.width - 1:
                        new_x += 1
                    else: pass
                
                    self.grid[x, y] = None
                    creature.x = new_x
                    creature.y = new_y
                    self.grid[new_x, new_y] = creature

        if iteration % self.harvesting_period == 0:
            # Harvesting phase
            self.harvest()

            # Replication phase
            n_creatures = len(self.get_creatures())
            if n_creatures < self.max_population:
                for _ in range(n_creatures):
                    x = random.randint(0, self.width - 1)
                    y = random.randint(0, self.height - 1)
                    parent = random.choice(self.get_creatures())
                    genome = parent.mutate(self.mutation_rate)
                    self.grid[x, y] = Creature(x, y, genome, self)

    def display(self):
        colors = ['#ffffff', '#000000', '#ff0000', '#00ff00', '#0000ff', '#ffff00', '#ff00ff', '#00ffff']
        genomes = [tuple(creature.genome) for creature in self.get_creatures()]
        color_dict = {genome: i for i, genome in enumerate(genomes)}
        color_dict[None] = 0
        color_grid = [[color_dict[tuple(creature.genome) if creature else None] for creature in row] for row in self.grid]
        color_grid = [[int(cell) for cell in row] for row in color_grid]
        cmap = ListedColormap(colors, 'indexed')
        plt.figure(figsize=(20,20))
        plt.imshow(color_grid, origin='lower', cmap=cmap)
        plt.show()

    def get_creatures(self):
        return [creature for row in self.grid for creature in row if creature is not None]

    def harvest(self):
        survival_distance = 10

        survivors = []
        for x in range(self.width):
            for y in range(self.height):
                creature = self.grid[x, y]
                if creature is not None and x <= survival_distance:
                    survivors.append(creature)
                    self.grid[x, y] = None
                else:
                    self.grid[x, y] = None
        random.shuffle(survivors)

        new_population = []
        while len(new_population) < self.max_population:
            parent1 = random.choice(survivors)
            parent2 = random.choice(survivors)
            child_genome = self.mix_genomes(parent1.genome, parent2.genome)
            child = Creature(parent1.x, parent1.y, child_genome, parent1.world)
            new_population.append(child)

        for idx, creature in enumerate(new_population):
            x, y = self.get_random_empty_position()
            creature.x = x
            creature.y = y
            self.grid[x, y] = creature






    def get_random_empty_position(self):
        while True:
            x = random.randint(0, self.width - 1)
            y = random.randint(0, self.height - 1)
            if self.grid[x, y] is None:
                return x, y

    def run(self, max_steps):
        self.populate(self.max_population)
        for iteration in range(max_steps):
            clear_output(wait=True)
            self.display()
            self.update(iteration)
            time.sleep(self.sleep_timer)

    def mix_genomes(self, genome1, genome2):
        child_genome = []
        for gene1, gene2 in zip(genome1, genome2):
            if random.choice([True, False]):
                child_genome.append(gene1)
            else:
                child_genome.append(gene2)
        return child_genome

world = World(128, 128, mutation_rate=0.1, harvesting_period=64, max_population=1000)
world.run(2000)
```
